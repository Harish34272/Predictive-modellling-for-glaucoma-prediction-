```python
import csv
import cv2
import sys
import numpy as np
from sklearn.decomposition import PCA
import os
from os import path
import sys
import numpy as np

hog = cv2.HOGDescriptor()
num_features=250
category=[]
hogarray=[]
base_dir = r"E:\study material\miniproject-2\Glaucoma-classification-ML-DL-main\Glaucoma-classification-ML-DL-main\dataset"

for folder in os.listdir(base_dir):
    folder_path = path.join(base_dir, folder)
    if path.isdir(folder_path):
        for filename in os.listdir(folder_path):
            print(filename)
            if filename[-3:] in ["jpg", "png"]:
                image = path.join(folder_path, filename)
                print(f'Reading image: {image}')  # Print out the image path
                img = cv2.imread(image)
                if img is None:
                    print(f'Failed to load image: {image}')  # Print out a message if the image failed to load
                    continue
                resized = cv2.resize(img, (64, 128), interpolation=cv2.INTER_AREA)
                h = hog.compute(resized)
                hogImage = h.T
                if folder == "Glaucoma":
                    category.append(1)
                else:
                    category.append(0)
                hogarray.append(hogImage)

if len(hogarray) == 0:
    print('No images were processed.')
else:
    hogarray_np=np.array(hogarray)
    reshaped_hog_array=np.reshape(hogarray_np,(hogarray_np.shape[0],hogarray_np.shape[1]))
    pca=PCA(n_components=num_features)
    reduced_features=pca.fit_transform(reshaped_hog_array)
    feature_list=reduced_features.tolist()
    csvdata=[]
    for id,line in enumerate(feature_list):
        newImg=line
        newImg.insert(0,category[id])
        csvdata.append(newImg)
    filename=str(num_features)+'extractedfeatures.csv'
    with open(filename,'w') as csvFile:
        writer=csv.writer(csvFile)
        writer.writerows(csvdata)
    csvFile.close()  # Corrected line
    print("Done extracting features")
```
